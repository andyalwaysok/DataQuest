Data: CIA World Factbook - Compendium of Stats about all countires on Earth

```python
%%capture
%load_ext sql
%sql sqlite:///factbook.db
```

Skimming data

```python
%%sql

SELECT *
FROM facts
LIMIT 5
```

Min & Max of population and growth

```python
%%sql

SELECT MIN(population), MAX(population), MIN(population_growth), MAX(population_growth)
FROM facts
```

Investigating which country has 0 and 7.2 bil population

```python
%%sql

SELECT *
FROM facts
WHERE population = 0
```

```python
%%sql

SELECT *
FROM facts
WHERE population = (SELECT MAX(population) FROM facts)
```

Min & Max of population and growth without World

```python
%%sql

SELECT MIN(population), MAX(population), MIN(population_growth), MAX(population_growth)
FROM facts
WHERE name != 'World'
```

Average population and area without World

```python
%%sql

SELECT AVG(population), AVG(area)
FROM facts
WHERE name != 'World'
```

Countries with most people and highest growth rate

```python
%%sql

SELECT name
FROM facts
WHERE population = (SELECT MAX(population) FROM facts WHERE name != 'World')
```

```python
%%sql

SELECT name
FROM facts
WHERE population_growth = (SELECT MAX(population_growth) FROM facts WHERE name != 'World')
```

Which countries have the highest ratios of water to land? 
Which countries have more water than land?

```python
%%sql

SELECT name, area_water, area_land, 
MAX((CAST(area_water AS FLOAT) / area_land) * 100) ratio
FROM facts
```

```python
%%sql

SELECT name, area_water, area_land, 
(CAST(area_water AS FLOAT) / area_land) * 100 ratio
FROM facts
WHERE ratio > 50
ORDER BY 4 DESC
```

Which countries will add the most people to their population next year?

```python
%%sql
SELECT name, population, population_growth, 
MAX(population * (population_growth/100)) growth
FROM facts
WHERE name != 'World'
ORDER BY 4 DESC
```

Which countries have a higher death rate than birth rate?

```python
%%sql

SELECT name, birth_rate, death_rate
FROM facts
WHERE birth_rate < death_rate
ORDER BY 3
```

What countries have the highest population/area ratio

```python
%%sql

SELECT name, population, area, population/area
FROM facts
ORDER BY 4 DESC
```
